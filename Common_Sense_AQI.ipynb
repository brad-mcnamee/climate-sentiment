For tweet scraping, I will be using snscrape, scraping hashtags related to air quality. I will be utilizing a list similar to Gurajala et al. (https://journals.sagepub.com/doi/full/10.1177/2056305119867656). The hashtags will include:
 #airpollution
 #airquality
 #airqualityindex
 #aqi
 #cleanair
 #ozone
 #smog

 #emissions

```python
import snscrape.modules.twitter as snt
import pandas as pd

# Using TwitterSearchScraper to scrape data and append tweets to list
#These tweets will be selected based on various criteria, including:
#lang - language
#timeframe - since:[date] until:[date]
#place - geolocation (only 1-2% of twitter users are likely to geotag, so better metric is based on user's profile)
# 

# Creating list to append tweet data to
airpollution_list = []
for i,tweet in enumerate(snt.TwitterSearchScraper('#airpollution lang:en since:2007-01-01 until:2021-12-31 near:NJ').get_items()):
    # if x tweets on this day, move to next day
    if i>1000000:
        break
    airpollution_list.append([tweet.date, tweet.content, tweet.user.username, tweet.id])
   
# Creating a dataframe from the tweets list above
airpollution_df = pd.DataFrame(airpollution_list, columns=['Datetime', 'Text', 'Username', 'Tweet Id'])

print(airpollution_df)
print(len(airpollution_list))
```

# Sentiment Analysis

Sentiment Analysis with VADER assigns a sentiment polarity score to each individual tweet 

```python
 from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
NJ_tweets['compound'] = [analyzer.polarity_scores(x)['compound'] for x in NJ_tweets['Text']]
NJ_tweets['neg'] = [analyzer.polarity_scores(x)['neg'] for x in NJ_tweets['Text']]
NJ_tweets['neu'] = [analyzer.polarity_scores(x)['neu'] for x in NJ_tweets['Text']]

NJ_tweets['pos'] = [analyzer.polarity_scores(x)['pos'] for x in NJ_tweets['Text']]
```

Different methods of sentiment scoring; first, is compound-sentiment (more accurate and even distribution)

```python
#This below method provides 'compound-sentiment', gaining the sentiment score based on the overall sentiment of the sentence
import numpy as np
conditions = [
    (NJ_tweets['compound'] >= 0.05),
    (NJ_tweets['compound'] > -0.05) & (NJ_tweets['compound'] < 0.05),
    (NJ_tweets['compound'] <= -0.05)
    ]

# create a list of the values we want to assign for each condition
values = ['pos', 'neu', 'neg']

# create a new column and use np.select to assign values to it using our lists as arguments
NJ_tweets['compound-sentiment'] = np.select(conditions, values)

NJ_tweets
```

```python
tokens = nlp(''.join(str(Neu_tweets.Text.tolist())))
```

```python
items = [x.text for x in tokens.ents]
Counter(items).most_common(20)
```

```python
#NER - people first
person_list = []

for ent in tokens.ents:
    if ent.label_ == 'PERSON':
        person_list.append(ent.text)
        
person_counts = Counter(person_list).most_common(20)
df_person = pd.DataFrame(person_counts, columns =['text', 'count'])
df_person
```

```python
norp_list = []
for ent in tokens.ents:
    if ent.label_ == 'NORP':
        norp_list.append(ent.text)
        
norp_counts = Counter(norp_list).most_common(20)
df_norp = pd.DataFrame(norp_counts, columns =['text', 'count'])
df_norp
```

```python
#sentiment wordcloud - ALL TWEETS
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

stopwords = STOPWORDS.copy()
stopwords.update(['http', 'https', 'co', 'starbuck', 'starbucks', 't', 'will']) # add some additional stopwords.
# make all the text lowercase and combine everything together.
all_txt = [txt.lower() for txt in NJ_tweets['Text'].to_list()]
all_txt = ' '.join(all_txt)

# create and plot the wordcloud.
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=800, height=600).generate(all_txt)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
